```python
import csv
import json
import gzip
import shutil
import os
import random
```

```python
def temp():
    videoNum = 10
    catDict = {
        "Autos & Vehicles": "A&V",
        "Comedy": "Com",
        "Education": "Edu",
        "Entertainment": "Ent",
        "Film & Animation": "F&A",
        "Gaming": "Gam",
        "Howto & Style": "H&S",
        "Music": "Mus",
        "Nonprofits & Activism": "N&A",
        "News & Politics": "N&P",
        "Pets & Animals": "P&A",
        "People & Blogs": "P&B",
        "Science & Technology": "S&T",
        "Sports": "Spo",
        "Travel & Events": "T&E"
    }

    A = 0
    C = 0
    Ed = 0
    En = 0
    F = 0
    G = 0
    H = 0
    M = 0
    Na = 0
    Np = 0
    Pa = 0
    Pb = 0
    St = 0
    Sp = 0
    T = 0

    while((A != 10) | (C != 10) | (Ed != 10) | (En != 10) | (F != 10) | (G != 10) | (H != 10) | (M != 10) | (Na != 10) | (Np != 10) | (Pa != 10) | (Pb != 10) | (St != 10) | (Sp != 10) | (T != 10)):
        file_number = random.randint(0, 1023)
        file_number_str = f"{file_number:04d}"

        # Build the input file name
        input_file_name = f"./../metadata/yttemporal1b_train_{file_number_str}of1024.jsonl.gz"
        # Build the temporary unzipped file name
        unzipped_file_name = f"./../metadata/yttemporal1b_train_{file_number_str}of1024.jsonl"

        # Extract the JSONL file from the gzip file
        with gzip.open(input_file_name, 'rb') as gz_file:
            with open(unzipped_file_name, 'wb') as json_file:
                shutil.copyfileobj(gz_file, json_file)


        # Read all the video data from the JSON Lines file into a list
        with open(unzipped_file_name, "r") as input_file:
            datas = [json.loads(line) for line in input_file]

        # Generate a random index to select a random video
        random_index = random.randint(0, len(datas) - 1)
        random_video_data = datas[random_index]

        category = random_video_data["categories"][0]
        category = catDict.get(category)
        count = 0
        if(category == "A&V"):
            count = A
        elif(category == "Com"):
            count = C
        elif(category == "Edu"):
            count = Ed
        elif(category == "Ent"):
            count = En
        elif(category == "F&A"):
            count = F
        elif(category == "Gam"):
            count = G
        elif(category == "H&S"):
            count = H
        elif(category == "Mus"):
            count = M
        elif(category == "N&A"):
            count = Na
        elif(category == "N&P"):
            count = Np
        elif(category == "P&A"):
            count = Pa
        elif(category == "P&B"):
            count = Pb
        elif(category == "S&T"):
            count = St
        elif(category == "Spo"):
            count = Sp
        elif(category == "T&E"):
            count = T

        output_file_path = f'''./../sampleData/{category}/dataNew.json'''
        if((random_video_data["duration"] > 600) & (count < 10)):
            with open(output_file_path, "a+") as output_file:
                formatted_data = json.dumps(random_video_data, indent=4)
                output_file.write(formatted_data)
            if category == "A&V":
                A += 1
            elif category == "Com":
                C += 1
            elif category == "Edu":
                Ed += 1
            elif category == "Ent":
                En += 1
            elif category == "F&A":
                F += 1
            elif category == "Gam":
                G += 1
            elif category == "H&S":
                H += 1
            elif category == "Mus":
                M += 1
            elif category == "N&A":
                Na += 1
            elif category == "N&P":
                Np += 1
            elif category == "P&A":
                Pa += 1
            elif category == "P&B":
                Pb += 1
            elif category == "S&T":
                St += 1
            elif category == "Spo":
                Sp += 1
            elif category == "T&E":
                T += 1

            os.remove(unzipped_file_name)


```
